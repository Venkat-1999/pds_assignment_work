```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



# Visualization 1: Bar Plot of Gender Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='gender', data=stp_data)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 2: Box Plot of Math Scores by Test Preparation Course
plt.figure(figsize=(8, 6))
sns.boxplot(x='test preparation course', y='math score', data=stp_data)
plt.title('Math Scores by Test Preparation Course')
plt.xlabel('Test Preparation Course')
plt.ylabel('Math Score')
plt.show()
```

```python
# Visualization 3: Box Plot of reading Scores by Test Preparation Course
plt.figure(figsize=(8, 6))
sns.boxplot(x='test preparation course', y='reading score', data=stp_data)
plt.title('Reading Scores by Test Preparation Course')
plt.xlabel('Test Preparation Course')
plt.ylabel('Reading Score')
plt.show()
```

```python
# Visualization 4: Box Plot of writing Scores by Test Preparation Course
plt.figure(figsize=(8, 6))
sns.boxplot(x='test preparation course', y='writing score', data=stp_data)
plt.title('Reading Scores by Test Preparation Course')
plt.xlabel('Test Preparation Course')
plt.ylabel('writing Score')
plt.show()
```

```python
# Visualization 5: Bar Plot of who done test preparation course
plt.figure(figsize=(8, 6))
sns.countplot(x='test preparation course', data=stp_data)
plt.title('Detailes of test preparation course')
plt.xlabel('test preparation')
plt.ylabel('Count')
plt.show()
```
