```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create a hypothetical dataset
fr_data = pd.DataFrame({
    'Frailty': ['N', 'N', 'N', 'Y', 'Y', 'N', 'Y', 'Y', 'N', 'N'],
    'Weight': [112, 136, 153, 142, 144, 123, 141, 136, 112, 120]
})

# Visualize categorical data using count plot
plt.figure(figsize=(10, 6))
sns.countplot(x='Frailty', data=fr_data)
plt.xlabel('Frailty')
plt.ylabel('members')
plt.title('Count Plot of Categorical Data')
plt.show()

# Visualize numerical data using histogram
plt.figure(figsize=(10, 6))
sns.histplot(fr_data['Weight'], bins=10, kde=True)
plt.xlabel('Weight')
plt.ylabel('Members')
plt.title('Histogram of Numeric Data')
plt.show()
```
